Bằng cách bật các ống đèn LED thích hợp ta có thể hiển thị chữ số bất kỳ. Số đèn sáng càng nhiều thì việc hiển thị chữ số đó càng tốn năng lượng. ## 2. Tư duy thuật toán
Khi nhìn vào bài toán, đầu tiên ta có thể thấy bài toán có điểm khá tương tự với bài toán Knapsack (bài toán điển hình của Quy Hoạch Động). Vậy nên ta có thể giải bài toán này với công thức quy hoạch động đơn giản:
>$A[n] = max (A[n - i] + u[i])$ với $0 \leq i \leq 9$

Tuy nhiên, nếu tiếp cận bài toán theo hướng này, độ phức tạp của giải thuật sẽ là $O(N * 9)$ $\rightarrow$ Sẽ dẫn đến TLE nếu giải theo phương thức này. Vậy nên chúng em đổi hướng tiếp cận qua giải thuật tham lam

## 3. Tìm kiếm quy luật
Phân tích bài toán, ta có thể nhận thấy sự tương quan giữa số bóng đèn led và số được hiển thị là: 

|    Số cần hiển thị    |   Số bóng đèn led   | 
|-----------------------|---------------------|
|         0             |          6          |
|         1             |          2          |
|         2             |          5          |
|         3             |          5          |
|         4             |          4          |
|         5             |          5          |
|         6             |          6          |
|         7             |          3          |
|         8             |          7          |
|         9             |          6          |

$\rightarrow$ Dễ nhận thấy, đây là một bài toán sử dụng giải thuật tham lam. Khi số bóng đèn chia hết cho 3, ta có thể lựa chọn số 7, khi số bóng đèn chia 3 dư 1, ta có thể lựa chọn số 7 và số 4. Khi bóng đèn chia 3 dư 2, ta có thể lựa chọn số 7 và số 1



## 4. Soucecode cho bài toán

```
n = int(input())

res, mod = (n // 3) * 7, n % 3
if (mod == 1):
    res = res - 7 + 4
elif (mod == 2):
    res += 1
print (res)
```
